1. What is the result of the code, and explain?
X = 'iNeuron'
def func():
print(X)
func()

Ans:- The output will be 'iNeuron'. The function func() is printing the global variable X, which is set to 'iNeuron'.

2. What is the result of the code, and explain?
X = 'iNeuron'
def func():
X = 'NI!'
func()
print(X)

Ans:- The output will be 'iNeuron'. Although X is set to 'NI!' inside the function func(), this assignment does not affect the global variable X. The print(X) statement after calling func() prints the global X, not the X inside func().

3. What does this code print, and why?
X = 'iNeuron'
def func():
X = 'NI'
print(X)
func()
print(X)

Ans:-The output will be 'NI' followed by 'iNeuron'. Inside the function func(), X is set to 'NI', and this value is printed. However, this does not affect the global variable X, which remains 'iNeuron'.

4. What output does this code produce? Why?
X = 'iNeuron'
def func():
global X
X = 'NI'
func()
print(X)

Ans:-The output will be 'NI'. The keyword global inside the function func() indicates that X refers to the global variable X. Therefore, when we set X to 'NI' inside func(), it changes the global variable X.

5. What about this code—what’s the output, and why?
X = 'iNeuron'
def func():
X = 'NI'
def nested():
print(X)
nested()
func()
X

Ans:-The output will be 'NI' followed by 'iNeuron'. The function nested() prints the value of X in its containing function’s scope, which is 'NI'. However, this does not affect the global variable X, which remains 'iNeuron'.

6. How about this code: what is its output in Python 3, and explain?

def func():
X = 'NI'
def nested():
nonlocal X
X = 'Spam'
nested()
print(X)
func()

Ans:- The output will be 'Spam'. The keyword nonlocal inside the function nested() indicates that X refers to the nearest enclosing scope that contains an X, excluding the global scope. Therefore, when we set X to 'Spam' inside nested(), it changes the X in its containing function’s scope.
